## Code to filter only the columns needed for zip code and citizenship

```python
import pandas as pd
```

```python
data = pd.read_csv('sdoh_ma.csv')
data.head()
```

```python
# ACS_PCT_CTZ_US_BORN_ZC: Percentage of population consisting of U.S. citizens born in United States, Puerto Rico, or U.S. Islands (ZCTA level)
# ACS_PCT_FOREIGN_BORN_ZC: Percentage of population that is foreign-born (ZCTA level)
```

```python
filtered_data = data[['ZIPCODE', 'STATE', 'REGION', 'ACS_PCT_CTZ_US_BORN_ZC', 'ACS_PCT_FOREIGN_BORN_ZC']]
filtered_data
```

```python
filtered_data['new_zipcode'] = filtered_data['ZIPCODE'].apply(lambda x: '0'+str(x) if len(str(x))==4 else 'x')
filtered_data = filtered_data.drop(columns='ZIPCODE')
```

```python
filtered_data.to_csv('filtered_data.csv', index=False)
```
